**1. Introduction**

The objective of this project is to classify authors based on their newspaper articles. The dataset consists of 1,500 articles written by 30 distinct authors (50 articles per author). The classification task involves multi-class text classification, and various natural language processing (NLP) techniques were explored.

**2. Data Processing**

To ensure efficient processing and model performance, the dataset underwent several preprocessing steps:

The dataset was loaded into a pandas DataFrame, where each row contained an author and their corresponding article.

The text data was tokenized for further analysis.

Text cleaning involved:

Converting all text to lowercase.

Removing punctuation marks.

Eliminating stopwords to retain only meaningful words.

Further filtering was applied by checking word frequency and removing outlier words that could distort contextual information.

**3. Word Embedding**

To transform textual data into numerical representations, we trained a Word2Vec model. This model allowed us to represent words as vectors in a high-dimensional space, capturing semantic relationships.

**4. Model Selection and Training**

We experimented with various models for classification, including:

GPT-2: A transformer-based language model for text generation and understanding.

BERT: A pre-trained transformer model known for its strong contextual understanding.

Neural Network: Deep learning with layers that creates features and learning from contex.

SVM (Support Vector Machine): A classical machine learning algorithm for classification tasks.

Gaussian Naive Bayes: A probabilistic model based on Bayes' theorem.

During model evaluation, 5-fold cross-validation was applied to ensure robust performance assessment.
